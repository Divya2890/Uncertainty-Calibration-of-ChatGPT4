```python
#### GETTING STATISTICS

import pandas as pd
from openpyxl import load_workbook
import re

def count_words(text):
    words = text.split()
    return len(words)

def count_sentences(text):
    # Split text into sentences using regex
    sentences = re.split(r'[.!?]+', text)
    # Remove empty strings (occurs if the last character is a punctuation mark)
    sentences = [sentence for sentence in sentences if sentence.strip()]
    return len(sentences)

# Load Excel file
file_path = 'Final_Random_Responses.xlsx'
workbook = load_workbook(file_path)
sheet_names = workbook.sheetnames

# Specify the column to analyze
column_to_analyze = 'Prompt_response_GPT_4'  # Change this to the column letter or name

# Process each sheet in the Excel file
for sheet_name in sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name)

    # Get the specified column
    column_data = df[column_to_analyze]

    # Initialize counters
    total_words = 0
    total_sentences = 0
    total_bold_words =0
    words_cell =[]
    sentence_cell=[]
    

    # Iterate through each cell in the specified column
    for cell in column_data:
        if isinstance(cell, str):
            # Count words and sentences
            total_words += count_words(cell)
            total_sentences += count_sentences(cell)
            words_cell.append(count_words(cell))
            sentence_cell.append(count_sentences(cell))
            # Check for bold formatting
            if '<b>' in cell:  # Modify this condition based on how bold is formatted in your Excel sheet
                total_bold_words += count_words(cell)

    # Output results for the current sheet
    print(f"Sheet: {sheet_name}")
    print(f"Total Words: {total_words}")
    print(f"Total Sentences: {total_sentences}")
    print(f"Total Bold Words: {total_bold_words}")
    print("setences in a cell",sentence_cell)
    print("words in a cell",words_cell)

```
